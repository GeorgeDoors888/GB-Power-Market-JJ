# NESO BSUoS Billing Report Analysis
This notebook analyzes the NESO BSUoS Billing Report JSON file to summarize key sections and extract insights.

```python
import json

# Load the JSON file
file_path = '../NESO_BSUoS_Billing_Report.json'
with open(file_path, 'r') as f:
    data = json.load(f)

# Display the report title and generated date
data['report_title'], data['generated_on']
```

```python
# Summarize the sections
for section in data['sections']:
    print(f"Section: {section['title']}")
    print(section['content'])
    print('-' * 80)
```

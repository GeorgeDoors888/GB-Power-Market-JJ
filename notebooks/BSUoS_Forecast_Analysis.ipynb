# BSUoS Forecast Analysis
This notebook analyzes the BSUoS forecast data to extract key metrics and generate visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '../neso_data_comprehensive/neso_portal/bsuos-monthly-forecast_BSUoS_Forecast_for_May_2020__Including_new_services.csv'
data = pd.read_csv(file_path)

# Display the first few rows
data.head()
```

```python
# Generate a line plot for Total Costs vs. Demand
plt.figure(figsize=(10, 6))
plt.plot(data['Month'], data['Total'], label='Total Costs', marker='o')
plt.plot(data['Month'], data['Demand'], label='Demand', marker='x')
plt.xlabel('Month')
plt.ylabel('Values')
plt.title('Total Costs vs. Demand Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Generate a bar plot for Profit/Loss
plt.figure(figsize=(10, 6))
plt.bar(data['Month'], data['Esitmated NGET Profit/(Loss)'], color='orange')
plt.xlabel('Month')
plt.ylabel('Profit/Loss (£m)')
plt.title('Profit/Loss Trends Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
